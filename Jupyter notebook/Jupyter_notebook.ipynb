### Импортируем библиотеку requests

```python
import requests
```

### Получаем результат сервиса

```python
# Параметр в запросе
param = {'count': '12'}
# Результат запроса
r = requests.get('http://127.0.0.1:5000/second_version', params=param)
# Убирает html-тег <h1><h1> и получаем строку
str = r.text[4:-4]
print(str)
```

### Преобразуем строку в лист значений чисел ряда Фибоначчи

```python
# Список чисел последовательности Фибоначчи
res = [int(elem) for elem in str.split(': ')[1].split(', ')]
```

## Визуализация

### Устанавливаем библиотеку matplotlib

```python
! pip install matplotlib
```

### Импортируем библиотеку matplotlib для визуализации

```python
import matplotlib.pyplot as plt
```

### Введем данные для осей графика

```python
x = [i+1 for i in range(len(res))] # номера
y = res # значения
```

### Строим столбчатую диаграмму

```python
fig = plt.figure(figsize = (10, 5)) # размер
plt.bar(x, y) # стобчатая диаграмма
plt.xlabel('Порядковый номер элемента ряда')
plt.ylabel('Значение элемента ряда')
plt.title(f'Первые {len(res)} чисел последовательности Фибоначчи')
# Определяем цену деления каждой оси
xy = plt.gca()
xy.xaxis.set_major_locator(plt.MultipleLocator(1))
xy.yaxis.set_major_locator(plt.MultipleLocator(10))
plt.show()
```

### Строим график

```python
fig = plt.figure(figsize = (10, 5)) # размер
plt.plot(x, y, c='red') # график
plt.xlabel('Порядковый номер элемента ряда')
plt.ylabel('Значение элемента ряда')
plt.title(f'Первые {len(res)} чисел последовательности Фибоначчи')
# Определяем цену деления каждой оси
xy = plt.gca()
xy.xaxis.set_major_locator(plt.MultipleLocator(1))
xy.yaxis.set_major_locator(plt.MultipleLocator(10))
plt.show()
```
